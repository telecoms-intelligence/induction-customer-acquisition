```python
# Is the data clean -
# Are there duplicate PJ_IDs?
unique_PD_ID = data.groupby('PJ_ID').agg({'pj_created_at': 'max'})
if (len(data) == len(unique_PD_ID)):
    print('PJ_ID is unique')
else:
    print('PJ_ID is not unique. Please clean data.')
```

```python
# Is the data clean -
# Is the link Id unique for every PJ if he clicked on the invitation email
unique_mpj = data.groupby('mpj').agg({'PJ_ID': 'nunique'})
if len(unique_mpj[unique_mpj['PJ_ID'] !=1]) == 0:
        print('mpj is unique')
else:
    print('mpj is not unique. Please clean data.')
```
